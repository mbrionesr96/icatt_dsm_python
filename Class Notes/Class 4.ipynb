There are five big question that Data Science tries to answer, conequently, there are five types of algorithms (sometimes they can be mixed tho).

## Machine Learning with Python 
Python has been set as the preferred language for ML due to its simplicity, readability, and extensive library support and its versatility. Some very important libraries for ML are: *NumPy, Pandas, Scikit-Learn, TensorFlow/PyTorch* and *Matplotlib/Seaborn*.

### Scikit-Learn
Is a library specifically built for ML and statistical modeling. It provides a set of ML algorithms such as classification, regression, clustering, dimensionality reduction, etc. 

#### Linear regresion
The **linear regresion algorithm** of Scikit-Learn tries to predict the behaviour of a dependent variable using the data of one or more independent variables. The algorithm realies on Ordinary Least Squares (OLS) method to fit the model to the data. Additional features include handling of missing values, converting categorical data to numerical data, and data normalization.

For a linear regression algorith we usually have the following steps: Loading Data, Data Exploration, Data Preprocessing, Train the Model, Make Predictions and Evaluation of the Model.

#### Logistic Regression

A **logistic regresion** predict binary outcomes, estimates a probability using a logistic function, it's suitable for binary and multiclass *classification* algorithms (not *regression* algorithms.

The model provided on Scikit-Learn handles both numerical and categorical data thanks to OvR (one vs rest) scheme. We can also access to the coefficients of the model and the model interpretation. 

The *model evaluation* is different from what we saw with linear regression, for example we may produce a confusion matrix for true vs. predicted classification.

This model can create four results, two of them would be errors: 
* TRUE negative: I predicted the i-th element belonging to the j-th category and it was indeed from j-th category.
* False negative: I predicted the i-th element belonging to the l-th category and it was indeed from j-th category.
* True Positives: I predicted the i-th element belonging to the j-th category and it was indeed from l-th category.
* False positives: I predicted the i-th element belonging to the l-th category and it was indeed from l-th category.

We can count how many elements there are on each category and from this numbers compute the accuracy, precision, recall, F1-score, ROC-AUC.
