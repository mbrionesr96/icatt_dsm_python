## Operators

On python we differentiate between Arithmetic, Assignment, Comparison and Logical Operators. 

### Arithmetic

The opertator **Modulus (%)** returns the remainder of **division (/)**

```python
10 / 6
```

```python
10 % 6
```

```python
#define a function which returns True if the number is even, False otherwise
def is_even(x): 
    result = x % 2
    if result==1:
        return False
    else: 
        return True

is_even(3)    
```

```python
is_even(4)
```

```python
#alternative function which returns True if the number is even, False otherwise
def is_even2(x): 
    if x % 2 == 0:
        return True
    
    return False

print(is_even2(3))
print(is_even2(6))
```

```python
#alternative function which returns True if the number is even, False otherwise
def is_even3(x): 
    return (x % 2 ==0)

print(is_even3(3))
print(is_even3(6))
```
