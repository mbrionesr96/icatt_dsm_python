### Arithmetic

The opertator **Modulus (%)** returns the remainder of **division (/)**

```python
10 / 6
```

```python
10 % 6
```

```python
#define a function which returns True if the number is even, False otherwise
def is_even(x): 
    result = x % 2
    if result==1:
        return False
    else: 
        return True

is_even(3)    
```

```python
is_even(4)
```

```python
#alternative function which returns True if the number is even, False otherwise
def is_even2(x): 
    if x % 2 == 0:
        return True
    
    return False

print(is_even2(3))
print(is_even2(6))
```

```python
#alternative function which returns True if the number is even, False otherwise
def is_even3(x): 
    return (x % 2 ==0)

print(is_even3(3))
print(is_even3(6))
```

## Data Structures
Values and variables are stored in structures like Lists, Tuples, Dictionaries and sets. 

### List
They are ordered, mutable collection of elements, the objects can be of different types. Created by **[]**

```python
my_list=[1,2,3,'python',4.5]
print(type(my_list))
print(my_list[4]) #remember that python counts from 0
```

```python
#moddify and add new elemet to the list
my_list[1] = 20
my_list.append(True)
print(my_list)
```

```python
#remove the element 3 from the list 
my_list.remove(my_list[2])
print(my_list)
```

### Tuples
Very similar to a list, we cannot modify after creation they are ideal for fixed data. Created by **()**

```python
my_tuple=("banana", "apple", "Cherry")
print(type(my_tuple))
print(my_tuple[0]) 
```

```python
my_tuple[0]="orange"
```

### Dictionaries
They are unordered (I cannot retrieve the first element because it doesn't exist) mutable collection of key-value pairs. Created by **{k1: v1, ...}**.

* `keys()`: returns a list of keys
* `values()`: returns a list of values
* `items()`: returns a list of key-values tuples

```python
my_dict={"name": "Maria", "age":29, "location": "Santiago"}
print(type(my_dict))
```

```python
print(my_dict[0]) #error because they are unordered
```

```python
print(my_dict["age"])
```

```python
#add another element
my_dict["country"]= "Chile"
my_dict
```

```python
# dictionaries are very flexible
my_dict["tuple"]=my_tuple
my_dict
```

### Sets
Unordered, mutable collection of *unique* elements. Created by **{}**. A set automatically removes duplicates. 

```python
my_set={1,2,3,4,5,6,3}
print(type(my_set))
print(my_set) #it eliminates the duplicated 3 
```

```python
# add an element
my_set.add(10)
print(my_set)
print(len(my_set))
```

## Control Flow Statements
Is the rder in which statements are executed. We regulate the control flow with conditional statemets and loops.

## Loops
A loop iterates a task, a *for* loop iterates over a sequence (list, tuple, string, etc)

```python
# for loop 
for i in range(1, 6): 
    print("I'm in the iteration "+str(i))
```

```python
#iterate over a list 
my_list=[1,4,7,10, -5]

for i in my_list: 
    print("I'm retrieving the element "+str(i))
```

```python
#iterate over a tuple
my_tuple=(3, -2, "hello", True)

for i in my_tuple: 
    print(f"I'm retrieving the element {i}")
```

In contrast a *while loop* iterates the task until the specified condition is True

```python
#set initial counter
counter=0

while counter<6:
    print(f"I'm starting the iterarion {counter}")
    counter += 1
    if counter == 4:
        break
    print(f"Closing in order to start the iterarion {counter}")

print(f"Final counter is {counter}")
```

As **break** we can also control the flow of the loop with **continue** or **pass**.

We must take care to specify conditions that will eventually become restrictive, aka **avoid infinite loops**

## Error Handling
Errors are managed by a special class of objects called *exceptions* we can use this class to prevent the user of the code we wrote. 
The most common errors are: *SytaxError, NameError, TypeError, IndexError, KeyError and ValueError*

```python
try : 
    x = 3 / 0

except ZeroDivisionError as e: 
    print(e)
```

```python
try: 
    x=10+"abc"
except TypeError as e: 
    print(f"The following error occured: {e}")
```

```python
# we can manage a generic error
try:
    my_list=[1,2,3,4,5]
    print(my_list[10])
except Exception as e: 
    print(f"The following error type has happened: {e}")
```
